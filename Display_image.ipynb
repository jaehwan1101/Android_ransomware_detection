```python
import os
import glob
import matplotlib.pyplot as plt
from PIL import Image
import random

# 이미지 출력 함수
def display_images(folder_path, title, num_images=1000):
    image_paths = glob.glob(os.path.join(folder_path, "*.png"))
    random.shuffle(image_paths)
    
    fig, axs = plt.subplots(10, 10, figsize=(20, 20))
    fig.suptitle(title, fontsize=16)
    
    for i, ax in enumerate(axs.flat):
        if i >= num_images:
            break
        image = Image.open(image_paths[i])
        ax.imshow(image, cmap="gray")
        ax.axis("off")
    
    plt.show()

# 주어진 경로의 Benign 및 Ransomware 이미지 출력
ransomware_path = r'C:\Users\pc\Desktop\CNN\Test_Image\Ransomware'
benign_path = r'C:\Users\pc\Desktop\CNN\Test_Image\Benign'

display_images(benign_path, "Benign Images")
display_images(ransomware_path, "Ransomware Images")
```
